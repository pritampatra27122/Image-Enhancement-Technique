# Image Enhancement Tehniques on CXR(Chest X-ray) Images

# 1. GAMMA CORRECTION

```python
import cv2
import numpy as np
```

```python
# Open the image.
img = cv2.imread('chest.jpg')
  
# Trying 4 gamma values.
for gamma in [3.5]:
      
    # Apply gamma correction.
    gamma_corrected = np.array(255*(img / 255) ** gamma, dtype = 'uint8')
  
    # Save edited images.
    cv2.imwrite('gamma_transformed'+str(gamma)+'.jpg', gamma_corrected)
```

# 2. BALANCE CONTRAST ENHANCEMENT TECHNIQUE

```python
import cv2
import numpy as np
```

```python
def balance_contrast(img):
    # Convert image to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # Calculate the mean and standard deviation of the image
    mean = np.mean(gray)
    std = np.std(gray)
    
    # Calculate the lower and upper bounds for the contrast
    lower = mean - std
    upper = mean + std
    
    # Clip the values of the image between the lower and upper bounds
    contrast_balanced = np.clip(gray, lower, upper)
    
    # Scale the clipped values back to the range [0, 255]
    contrast_balanced = cv2.normalize(contrast_balanced, None, 0, 255, cv2.NORM_MINMAX)
    
    # Convert the result back to an 8-bit integer
    contrast_balanced = np.uint8(contrast_balanced)
    
    # Return the balanced image
    return contrast_balanced
```

```python
# Load an image
img = cv2.imread("chest.jpg")

# Balance the contrast
balanced_img = balance_contrast(img)

# Show the original and balanced images
cv2.imshow("Original", img)
cv2.imshow("Contrast Balanced", balanced_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# 3. Contrast Limited Adaptive Histogram Equilization

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def clahe(img, clipLimit=2.0, tileGridSize=(8, 8)):
    # Convert image to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # Create the CLAHE object
    clahe = cv2.createCLAHE(clipLimit=clipLimit, tileGridSize=tileGridSize)
    
    # Apply CLAHE to the grayscale image
    clahe_img = clahe.apply(gray)
    
    # Return the CLAHE result
    return clahe_img
```

```python
# Load an image
img = cv2.imread("chest.jpg")

# Apply CLAHE
clahe_img = clahe(img)
```

```python
# Show the original and CLAHE images
cv2.imshow("Original", img)
cv2.imshow("CLAHE", clahe_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Plot the original and CLAHE images
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
ax1.set_title("Original")
ax2.imshow(clahe_img, cmap="gray")
ax2.set_title("CLAHE")
plt.show()
```
