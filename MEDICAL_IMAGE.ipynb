# Implementation of Image Enhancement Techniques on CT and MRI Images 

# 1. Image Binarization

# 1.1 Importing Necessary Packages

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

# 1.2 Function to Read the Image

```python
def read_this(image_file, gray_scale=False):
    image_src = cv2.imread(image_file)
    if gray_scale:
        image_src = cv2.cvtColor(image_src, cv2.COLOR_BGR2GRAY)
    else:
        image_src = cv2.cvtColor(image_src, cv2.COLOR_BGR2RGB)
    return image_src
```

# 1.3 Function to Revalue the pixels which are less than 255

```python
def convert_binary(image_matrix, thresh_val):
    white = 255
    black = 0
    
    initial_conv = np.where((image_matrix <= thresh_val), image_matrix, white)
    final_conv = np.where((initial_conv > thresh_val), initial_conv, black)
    
    return final_conv
```

# 1.4 Binarizing Function

```python
def binarize_this(image_file, thresh_val=105, with_plot=False, gray_scale=False):
    image_src = read_this(image_file=image_file, gray_scale=gray_scale)
    
    if not gray_scale:
        cmap_val = None
        r_img, g_img, b_img = image_src[:, :, 0], image_src[:, :, 1], image_src[:, :, 2]
        
        r_b = convert_binary(image_matrix=r_img, thresh_val=thresh_val)
        g_b = convert_binary(image_matrix=g_img, thresh_val=thresh_val)
        b_b = convert_binary(image_matrix=b_img, thresh_val=thresh_val)
        
        image_b = np.dstack(tup=(r_b, g_b, b_b))
    else:
        cmap_val = 'gray'
        image_b = convert_binary(image_matrix=image_src, thresh_val=thresh_val)
    
    if with_plot:
        fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10, 20))
        
        ax1.axis("off")
        ax1.title.set_text('Normal Brain MR Image')
        
        ax2.axis("off")
        ax2.title.set_text("Binarized MR Image")
        
        ax1.imshow(image_src, cmap=cmap_val)
        ax2.imshow(image_b, cmap=cmap_val)
        return True
    return image_b
```

# 1.5 OUTPUT: Binarization of Brain MR Image

```python
binarize_this(image_file='Brain_MR.jpg', with_plot=True, gray_scale=True)
```

# 2. Contrast Limited Adaptive Histogram Equilization(CLAHE)

# 2.1. Importing Necessary Packages

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# 2.2 CLAHE Function 

```python
def clahe(img, clipLimit=5.0, tileGridSize=(8, 8)):
    # Convert image to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    clahe = cv2.createCLAHE(clipLimit=clipLimit, tileGridSize=tileGridSize)
    clahe_img = clahe.apply(gray)
    return clahe_img
```

# 2.3 Applying CLAHE on Brain CT Image

```python
img1 = cv2.imread("Brain_CT.jpg")
clahe_img1 = clahe(img1)
# Show the original and CLAHE images
cv2.imshow("Original Brain CT Image", img1)
cv2.imshow("CLAHE Brain CT Image", clahe_img1)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Plot the original and CLAHE images
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
ax1.set_title("Original CT Image")
ax2.imshow(clahe_img1, cmap="gray")
ax2.set_title("CLAHE CT Image")
plt.show()
```

# 2.4 Applying CLAHE on Brain MR Image

```python
# Load an image
img2 = cv2.imread("Brain_MR.jpg")
clahe_img2 = clahe(img2)
# Show the original and CLAHE images
cv2.imshow("Original Brain MR Image", img2)
cv2.imshow("CLAHE Brain MR Image", clahe_img2)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Plot the original and CLAHE images
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
ax1.set_title("Original MR Image")
ax2.imshow(clahe_img2, cmap="gray")
ax2.set_title("CLAHE MR Image")
plt.show()
```

# 3. Contrast Stretching

# 3.1 Importing Necessary Packages

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# 3.2 Contrast Stretching Function

```python
def contrast_stretch(img, lower_percentile=2, upper_percentile=98):
    # Convert the image to float type
    img = np.float32(img)
    # Calculate the lower and upper bounds
    lower_bound = np.percentile(img, lower_percentile)
    upper_bound = np.percentile(img, upper_percentile)
    # Stretch the contrast of the image
    img = np.clip(255 * (img - lower_bound) / (upper_bound - lower_bound), 0, 255)
    # Convert the image back to 8-bit unsigned integer type
    img = np.uint8(img)
    return img
```

# 3.3 Applying Contrast Stretching on brain CT Image

```python
# Load the image
img = cv2.imread("Brain_CT.jpg", cv2.IMREAD_GRAYSCALE)
# Perform contrast stretching
result = contrast_stretch(img)
# Plot the original and the contrast-stretched images
plt.figure()
plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis('on')

plt.subplot(122)
plt.imshow(result, cmap='gray')
plt.title("Contrast-stretched CT Image")
plt.axis('on')
plt.show()
```

```python
# Load the image
img = cv2.imread("Brain_MR.jpg", cv2.IMREAD_GRAYSCALE)
# Perform contrast stretching
result = contrast_stretch(img)
# Plot the original and the contrast-stretched images
plt.figure()
plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis('on')

plt.subplot(122)
plt.imshow(result, cmap='gray')
plt.title("Contrast-stretched MR Image")
plt.axis('on')
plt.show()
```
